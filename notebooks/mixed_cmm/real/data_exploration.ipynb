```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

This notebook serves as a preliminary exploration of the TB data. 

### Dataset overview

```python
tb_df = pd.read_csv('../../../data/real/processed/tb_pheno_geno_clean.csv')

tb_df.columns = tb_df.columns.str.strip()
outcome_cols = ['interp_bdq12', 'interp_bdq25', 
                'interp_dlm016', 'interp_dlm06', 
                'interp_lnz1']
for col in outcome_cols:
    tb_df[col] = tb_df[col].astype('Int64')

tb_df.head()
```

```python
print(f"Shape: {tb_df.shape}")
print(f"\nColumn types:\n{tb_df.dtypes.value_counts()}")
```

```python
#missing values
missing = tb_df.isnull().sum()
print(f"Missing values:\n{missing[missing > 0].sort_values(ascending=False)}")
```

Delamanid has the most missing data (12 rows for dlm016, 9 for dlm06). `lnz_mic` has 1 missing value.

```python
#missing data pattern for delamanid
missing_mask = tb_df['interp_dlm016'].isna()
print(f"Missing DLM rows: {missing_mask.sum()}")
print(pd.crosstab(tb_df.loc[missing_mask, 'lineage'], tb_df.loc[missing_mask, 'type']))
#if concentrated in one lineage or type, missingness is not at random and could bias the model
```

Missingness is roughly proportional to group size across lineages and resistance types, suggesting missing at random (MAR). No strong systematic pattern.

```python
#lineage and resistance type
print(tb_df['lineage'].value_counts())
print()
print(tb_df['type'].value_counts())
```

```python
#lineage × resistance type cross-tab
print(pd.crosstab(tb_df['lineage'], tb_df['type'], margins=True, margins_name='total'))
#lineage 4 and type 0 (MDR) dominate; lineage-specific mutations could confound resistance associations
```

Lineage 4 is almost exclusively MDR (93/100 strains, type 0), while lineage 2 has more diverse resistance types (36 MDR, 15 preXDR, 6 XDR). Since mutations can be lineage-specific, lineage and resistance type are not independent, which could confound mutation-resistance associations.

### Outcome variables

```python
fig, axes = plt.subplots(1, 5, figsize=(15, 4))
for ax, col in zip(axes, outcome_cols):
    tb_df[col].value_counts().sort_index().plot(kind='bar', ax=ax, color=['steelblue', 'red'])
    ax.set_title(col)
    ax.set_xlabel('0=S, 2=R')
    ax.set_ylabel('Patients')
    ax.tick_params(axis='x', rotation=0)
plt.suptitle('Resistance interpretation per drug/concentration')
plt.tight_layout()
plt.show()
```

Strong class imbalance across all drugs: most strains are susceptible. LNZ has only 1 resistant strain. BDQ at 0.25 mg/L has 1 resistant strain. DLM has the most resistant cases (~38 at dlm016), which makes it the most tractable outcome for modeling.

```python
#MIC distributions
print(tb_df[['bdq_mic', 'cfz_mic', 'lnz_mic', 'dlm_mic', 'ptm_mic']].describe())
```

```python
cc = {'bdq_mic': 1, 'cfz_mic': 1, 'lnz_mic': 1, 'dlm_mic': 0.06, 'ptm_mic': 1}

fig, axes = plt.subplots(1, 5, figsize=(15, 4))
for ax, col in zip(axes, ['bdq_mic', 'cfz_mic', 'lnz_mic', 'dlm_mic', 'ptm_mic']):
    counts = tb_df[col].dropna().value_counts().sort_index()
    counts.plot(kind='bar', ax=ax, color='steelblue')
    # add CC line
    labels = [float(t.get_text()) for t in ax.get_xticklabels()]
    if cc[col] in labels:
        idx = labels.index(cc[col])
        ax.axvline(x=idx - 0.5, color='red', linestyle='--', label='CC')
    ax.set_title(col)
    ax.set_ylabel('Patients')
    ax.set_xlabel('MIC (mg/L)')
    ax.tick_params(axis='x', rotation=45)
    ax.legend()

plt.suptitle('MIC distributions (red line = critical concentration (resistant if MIC > CC))')
plt.tight_layout()
plt.show()
```

Most strains have MICs well below the critical concentration for all drugs. BDQ and PTM have a small number of strains above CC (max=2 mg/L). DLM MICs are tightly clustered near 0.016 mg/L, far below CC=0.06.

```python
#outcome consistency: agar proportion interpretation vs MIC-based resistance
cc = {'bdq': 1, 'dlm': 0.06, 'lnz': 1}
pairs = [
    ('interp_bdq12',  'bdq_mic', 'bdq'),
    ('interp_bdq25',  'bdq_mic', 'bdq'),
    ('interp_dlm016', 'dlm_mic', 'dlm'),
    ('interp_dlm06',  'dlm_mic', 'dlm'),
    ('interp_lnz1',   'lnz_mic', 'lnz'),
]
fig, axes = plt.subplots(1, 5, figsize=(15, 4))
for ax, (interp_col, mic_col, drug) in zip(axes, pairs):
    mic_res = (tb_df[mic_col] > cc[drug]).map({True: 'R (MIC)', False: 'S (MIC)'})
    ct = pd.crosstab(tb_df[interp_col].map({0: 'S', 2: 'R'}), mic_res)
    ct = ct.reindex(index=['S', 'R'], columns=['S (MIC)', 'R (MIC)'], fill_value=0)
    sns.heatmap(ct, annot=True, fmt='d', cmap='Blues', ax=ax, cbar=False)
    ax.set_title(f'{interp_col}\nvs {mic_col}')
    ax.set_xlabel('MIC-based')
    ax.set_ylabel('Agar interp')
plt.suptitle('Outcome consistency: agar proportion interpretation vs MIC-based resistance')
plt.tight_layout()
plt.show()
```

BDQ and LNZ show good agreement between agar and MIC methods. DLM however shows major disagreement: 32 strains (dlm016) and 22 strains (dlm06) are classified as R by agar but S by MIC.

### Mutation features

```python
#Mutation sparsity
mut_cols = [c for c in tb_df.columns if c.startswith(('rv', 'mmpl', 'mmps', 'atpe', 'pepq', 'rplc', 'rrl', 'fbia', 'fbib', 'fbic', 'fbid', 'fgd1', 'ddn'))]
print(f"Mutation columns: {len(mut_cols)}")
prevalence = tb_df[mut_cols].mean() * 100
print(f"\nPrevalence stats:\n{prevalence.describe()}")
print(f"\nMutations in <5 patients: {(tb_df[mut_cols].sum() < 5).sum()}")
print(f"Mutations in 0 patients: {(tb_df[mut_cols].sum() == 0).sum()}")
```

```python
mutation_cols = tb_df.loc[:, 'rv0678_Arg140_Tyr145dup':].columns
prevalent = (tb_df[mutation_cols].mean() * 100)
print(prevalent[prevalent >= 5].sort_values(ascending=False))
```

```python
print(tb_df[['rv1979c_A-129G', 'mmpl5_Ile948Val']].value_counts()) #near-universal mutations
```

The feature space is extremely sparse: 170/193 mutations appear in fewer than 5 patients (mean prevalence 3.4%). Two mutations are near-universal (`rv1979c_A-129G` at 98.8%, `mmpl5_Ile948Val` at 98.2%) and carry no discriminative information. Only 14 mutations exceed 5% prevalence and are viable candidates for modeling.

```python
#co-occurrence of prevalent mutations
prevalent_muts = prevalent[(prevalent >= 5) & (prevalent < 98)].index.tolist()
fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(tb_df[prevalent_muts].corr(), cmap='coolwarm', center=0, ax=ax, square=True)
ax.set_title('Mutation co-occurrence (Pearson correlation)')
ax.tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
#mutations within the same gene tend to co-occur; strong clusters may collapse to a single feature in selection
```

Clear correlation clusters exist within genes (mmpl5, fgd1, fbic variants). `rv0678_G*128C` also correlates strongly with both mmpl5 mutations across gene boundaries. `rrl_1476188_C>T` and the pepq mutations are largely independent. 

```python
#co-occurrence clusters at |r| > 0.5 (connected components on the thresholded correlation graph)
import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import connected_components

corr = tb_df[prevalent_muts].corr().abs().values
adj = (corr > 0.5) & ~np.eye(len(prevalent_muts), dtype=bool)
_, labels = connected_components(csr_matrix(adj), directed=False)

groups = {}
for mut, lab in zip(prevalent_muts, labels):
    groups.setdefault(lab, []).append(mut)
for i, members in enumerate(sorted(groups.values(), key=lambda c: -len(c))):
    tag = f'cluster {i+1}' if len(members) > 1 else 'singleton'
    print(f'{tag} (n={len(members)}): {members}')
```
